```python
import json
import glob
import re
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
from collections import Counter

import pandas as pd
```

```python
data = []

re_attacker = re.compile("jiat75", flags=re.IGNORECASE)
re_main = re.compile("Larhzu", flags=re.IGNORECASE)

events = {
    'attacker' : [],
    'main' : [],
    'refers attacker' : [],
    'refers main': []
}

logins = Counter()

for g in glob.glob("*.txt"):
    with open(g) as f:
        for line in f:
            jevent = json.loads(line)
            #emails[jevent['actor']['email']] += 1
            logins[jevent['actor']['login']]+=1
            if re_attacker.search(line):
                if re_attacker.search(jevent['actor']['login']):
                    events['attacker'].append(jevent)
                else:
                    events['refers attacker'].append(jevent)
            if re_main.search(line): # check both conditions (this and above) independently
                if re_main.search(jevent['actor']['login']):
                    events['main'].append(jevent)
                else:
                    events['refers main'].append(jevent)
```

```python
def to_dataset(events, filter=None):
    dates = []
    for name in events:
        for e in events[name]:
            d = datetime.fromisoformat(e['created_at'])
            dates.append({'date' : d, 'event' : e['type'], 'who': name, 'repo':e['repo']['name'], 'actor':e['actor']['login']})
    
    df = pd.DataFrame(dates)

    fdf = df
    if filter:
        fdf = filter(df)
    
    return fdf

def plot(fdf, x = 'date', col='who', row='event'):
    sns.displot(fdf, x=x, col=col, row=row,)
    plt.savefig('project_xz_activities.pdf')

    #sns.displot(fdf, x='date', hue='who', row='event', multiple="dodge",)
    #sns.displot(fdf, x='date', row='who', hue='event', multiple="dodge",)

    #for ax in plt.gcf().axes:
    #    plt.sca(ax)
    #    plt.xticks(rotation=30)


def plot2(fdf):
    
    fdf = fdf[~fdf['who'].str.contains('refers')]
    fdf['M'] = fdf['date'].dt.strftime('%Y-%m')
    g = fdf.groupby(['event', 'who', 'M']).count().reset_index()
    g.loc[g['who']=='attacker', 'T'] = -1
    g.loc[g['who']!='attacker', 'T'] = 1
    g['D'] = g['T'] *g['date']
    #g.apply(lambda x: print(x[['who']]))
    for e in g['event'].unique():
        display(e)
        D = g[(g['event']==e) & (g['who']!='attacker')].sort_values('M')
        plt.bar(x=D['M'], height=D['D'])
        D = g[(g['event']==e) & (g['who']=='attacker')].sort_values('M')
        plt.bar(x=D['M'], height=D['D'])
        plt.show()
    display(g)
```

```python
def date_filter(df):
    fdf = df[(df['date'] <= datetime(2024,3,29,0,0).astimezone()) & (df['repo'].str.contains('tukaani-project/xz'))]
    return fdf

df = to_dataset(events, date_filter)

df['M'] = df['date'].dt.strftime('%Y-%m')
#df['event'].unique()

df['Interaction'] = False

for e in ['IssueCommentEvent', 'PullRequestReviewEvent', 'CommitCommentEvent']:
    df['Interaction'] |=  (df['event'] == e)

A = df[df['Interaction']].groupby('M').count().reset_index()
B = df[df['Interaction'] & ( df['who'] == 'attacker')].groupby('M').count().reset_index()
C = df[df['Interaction'] & ( df['who'] == 'main')].groupby('M').count().reset_index()
#plot2(df)
C
```

```python
df_all = A[['M','who']].merge(B[['M', 'who']], on='M', how="outer")
df_all = df_all.merge(C[['M','who']], on='M', how="outer")
```

```python
df_all = df_all.fillna(0)
df_all.columns = ['M', 'all', 'attacker', 'main']
df_all['attacker%'] = df_all['attacker'] / df_all['all']
df_all['main%'] = df_all['main'] / df_all['all']
```

```python
df_all.to_csv("../../scripts/git_logs/xz/interactions.csv", sep=';')
```

```python
fig, axes = plt.subplots(2, figsize=(10, 2), sharex=True)
axes[0].bar(df_all['M'], df_all['all'], width=0.5)

axes[1].bar(df_all['M'], df_all['main'], width=0.5)
axes[1].bar(df_all['M'], -1 * df_all['attacker'], width=0.5)

sns.despine()
```
